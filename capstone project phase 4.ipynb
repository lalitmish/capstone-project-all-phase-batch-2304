# 4.1 "Create 3 clusters as below.

1.Sensitive Area's

2.Moderate Area's

3.Peaceful Area's"

```python
df1=pd.read_csv("District_wise_crime_from_2001_2012.csv")
df1
```

```python
df_Total_CRIME = df1.sort_values(by='Sum of TOTAL IPC CRIMES', ascending=False)
df_Total_CRIME
```

We observed that DELHI UT districts is at top position in crime with  633174 crime and RAILWAYS KATRA is at lowest postion with
total 2 crime  from 2001 to 2012 out of 807 districs

```python
df3 = df_Total_CRIME[['Sr.no', 'Sum of TOTAL IPC CRIMES']]
df3.head
```

```python
#elbow method
from sklearn.cluster import KMeans

wcss=[]
for i in range (1,4):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df3)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,4),wcss)
plt.title('Sensitive,Moderate,Peaceful areas')
plt.xlabel('Number of cluster for')
plt.ylabel('Type of area')
plt.show()
```

```python
# fitting K-means to dataset
kmeans = KMeans(n_clusters =3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(df3)
print(y_kmeans)
```

We divide all data in three cluster 
1.Sensitive Area's   denoted by cluster 1

2.Moderate Area's    denoted by cluster 2

3.Peaceful Area's"   denoted by cluster 0

```python
# It predict the cluster number to which the datapoint belong to
test=kmeans.predict(np.asarray([[193,633174]]))
test[0]
```

it means that it come in sensitive area

```python
# It predict the cluster number to which the datapoint belong to
test1=kmeans.predict(np.asarray([[608,2]]))
test1[0]
```

it means that it come in peaseful area

```python
# It predict the cluster number to which the datapoint belong to
test3=kmeans.predict(np.asarray([[62,350347]]))
test3[0]
```

it means that it come in sensitive area
