```python
df=pd.read_csv("different type of crime rate statewise with literacy and population from 2001 to 2012.csv")

df
```

2.1 Analysis of Literacy Rate vs Total Crimes.

```python
#visualization
plt.bar(df['Literacy Rate'], df['TOTAL CRIMES'])
plt.xlabel('Literacy Rate')
plt.ylabel('TOTAL CRIMES')
plt.show()
```

We observe that at literacy rate 82 in that state total crime is highest and at literacy rate 65,70,74,81,86,89,92 total crime 
lowest percentage.  hence we can observe that if literacy rate is high then total crime is low it will not sure for every state
in some state literacy rate is low but total crime is also low

2.2 Analysis of the type of crime vs each state vs Literacy rate.
